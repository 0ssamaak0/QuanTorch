```python
from sankeyflow import Sankey
import matplotlib.pyplot as plt
```

```python
nodes = [
    [('Internships', 15)],
    [('No Reply', 8), ("Interview", 3, {"color": "yellow"}), ("Accepted", 2, {"color": "green"}), ('Rejected', 2, {"color": "red"}),],
    [('No Reply.', 9, {"label": "No Reply"}), ("Accepted.", 4, {'color': 'green', "label": "Accepted"})],
    [("Turned Down", 3), ("Agreed", 1)]
]
flows = [
    ('Internships', 'No Reply', 8),
    ('Internships', 'Interview', 3),
    ('Internships', 'Accepted', 2), 
    ('Internships', 'Rejected', 2),


    ('No Reply', 'No Reply.', 8), 
    ('Interview', 'No Reply.', 1), 
    ("Interview", "Accepted.", 2),
    ("Accepted", "Accepted.", 2),

    ('Accepted.', "Turned Down", 3),
    ('Accepted.', "Agreed", 1)
] 


s = Sankey(flows=flows, nodes=nodes, flow_opts=dict(curvature=0.3), node_opts=dict(label_format='{label} ({value})'))
s.draw()
```
